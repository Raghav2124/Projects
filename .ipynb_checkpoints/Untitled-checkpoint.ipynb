```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
data['EFFECTIVE_YR'] = pd.to_numeric(data['EFFECTIVE_YR'], errors='coerce')
```

```python
columns_to_impute = ['EFFECTIVE_YR', 'PREMIUM', 'PROD_YEAR','SEATS_NUM','CARRYING_CAPACITY']

# Create a subset of the DataFrame containing only the columns to impute
data_subset = data[columns_to_impute]

# Create an IterativeImputer object
imputer = IterativeImputer(max_iter=10 ,random_state=0)

# Impute missing values
data_imputed = imputer.fit_transform(data_subset)

# Convert the result back to a DataFrame
data[columns_to_impute] = pd.DataFrame(data_imputed, columns=columns_to_impute)
```

```python
data.loc[:, 'INSURED_VALUE'] = data['INSURED_VALUE'].astype(int)
data.loc[:, 'PREMIUM'] = data['PREMIUM'].astype(int)
data.loc[:, 'PROD_YEAR'] = data['PROD_YEAR'].astype(int)
data.loc[:, 'SEATS_NUM'] = data['SEATS_NUM'].astype(int)
data.loc[:, 'CCM_TON'] = data['CCM_TON'].astype(int)
```

```python
def remove_outliers(data):
    # Calculate IQR
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers = ((data < lower_bound) | (data > upper_bound)).any(axis=1)
    
    # Remove outliers
    data_cleaned = data[~outliers]
    
    return data_cleaned

# Apply the function to your DataFrame
data_cleaned = remove_outliers(data)
```
